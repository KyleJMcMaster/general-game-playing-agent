Generates data for training model

in format `input_board;move;output_board\n`

```python
import random
import numpy as np
num_lines = 100000
filename = '2048_move_data.txt'


def generate_random_board():
    tiles = [i for i in range(0, 15)]
    weights = [1 for i in range(0,15)]
    tiles = random.choices(population=tiles, weights=weights, k=16)
    out = ''
    for tile in tiles:
        if tile < 10:
            out += "0"+str(tile)
        else:
            out += str(tile)
    return out

def count_digits(n):
    count = 1
    while n // 10 >= 1:
        n /= 10 
        count += 1
    return count

def print_state(s):

    enc_state = [2**int(''.join(s[i:i+2])) for i in range(0, len(s), 2)]
    for i, tile in enumerate(enc_state):
        print(' ' * max(0, 5 - count_digits(tile)) + (str(tile) if tile != 1 else '0'), end=('\n' if (i+1) % 4 == 0 else ''))

    print('\n')

def apply_move(s, move):
        # plays move
        # 0:r,1:l,2:u,3:d
        s = [2**int(''.join(s[i:i+2])) for i in range(0, len(s), 2)]
        for i in range(16):
            if s[i] == 1:
                s[i] = 0

        if move == 0:
            for k in range(0, 3):
                for i in range(3 - k, 16 - k, 4):
                    for j in range(1, 4 - k):
                        if s[i] == 0 and s[i - j] != 0:
                            s[i] = s[i - j]
                            s[i - j] = 0
                        if s[i] != 0 and s[i - j] == s[i]:
                            s[i] *= 2
                            s[i - j] = 0
                            break
                        if s[i] != 0 and s[i - j] != 0 and s[i - j] != s[i]:
                            break
        if move == 1:
            for k in range(0, 3):
                for i in range(0 + k, 13 + k, 4):
                    for j in range(1, 4 - k):
                        if s[i] == 0 and s[i + j] != 0:
                            s[i] = s[i + j]
                            s[i + j] = 0
                        if s[i] != 0 and s[i + j] == s[i]:
                            s[i] *= 2
                            s[i + j] = 0
                            break
                        if s[i] != 0 and s[i + j] != 0 and s[i + j] != s[i]:
                            break
        if move == 2:
            for k in range(0, 3):
                for i in range(0 + 4 * k, 4 + 4 * k):
                    for j in range(1, 4 - k):
                        if s[i] == 0 and s[i + j * 4] != 0:
                            s[i] = s[i + j * 4]
                            s[i + j * 4] = 0
                        if s[i] != 0 and s[i + j * 4] == s[i]:
                            s[i] *= 2
                            s[i + j * 4] = 0
                            break
                        if s[i] != 0 and s[i + j * 4] != 0 and s[i + j * 4] != \
                                s[i]:
                            break
        if move == 3:
            for k in range(0, 3):
                for i in range(12 - 4 * k, 16 - 4 * k):
                    for j in range(1, 4 - k):
                        if s[i] == 0 and s[i - j * 4] != 0:
                            s[i] = s[i - j * 4]
                            s[i - j * 4] = 0
                        if s[i] != 0 and s[i - j * 4] == s[i]:
                            s[i] *= 2
                            s[i - j * 4] = 0
                            break
                        if s[i] != 0 and s[i - j * 4] != 0 and s[i - j * 4] != \
                                s[i]:
                            break
        
        # set new tile
        empty_squares = []
        for i in range(16):
            if s[i] == 0:
                empty_squares.append(i)
        
        if len(empty_squares)>0:
            pos = random.sample(empty_squares, 1)[0]
            val = random.choices([2, 4], weights=[0.9, 0.1], k=1)[0]
            s[pos] = val

        for i in range(16):
            if s[i] == 0:
                s[i] = 1
        s = [np.log2(s[i]).astype(int) for i in range(0, 16)]
        out = ''
        for tile in s:
            if tile < 10:
                out += "0"+str(tile)
            else:
                out += str(tile)
        return out


def generate_step():
    board = generate_random_board()
    move = random.choice([0,1,2,3])
    result = apply_move(board, move)
    dirs = ['R', 'L', 'U', 'D']
    return board+';'+dirs[move]+';'+result

with open(filename, 'a+') as f:
    for line in range(num_lines):
        f.write(generate_step()+'\n')

# dirs = ['Right', 'Left', 'Up', 'Down']
# for _ in range(20):
#     res = generate_step()
#     b,m,r = res.split(';')
#     print_state(b)
#     print(dirs[int(m)])
#     print_state(r)
```
